# Прогноз музыкального жанра
_____
**Описание проекта:**

Популярный музыкальный стриминговый сервис "МиФаСоль" расширяет работу с новыми артистами и музыкантами, в связи с чем возникла задача - правильно классифицировать новые музыкальные треки, чтобы улучшить работу рекомендательной системы. Сотрудники отдела работы со звуком подготовили датасет, в котором собраны некоторые характеристики музыкальных произведений и их жанры. Основная задача - разработать модель, позволяющую классифицировать музыкальные произведения по жанрам и получить предсказания для тестовой выборки.

**Метрика:**

Оценку качества алгоритма проводим по метрике "F1-micro".

**Описание данных:**

- **train.csv** - информация (~20000) музыкальных треках, которые будут использоваться в качестве обучающих данных.
- **test.csv** - информация (~5000) музыкальных треках, которые будут использоваться в качестве тестовых данных. Задача - предсказать значение 'music_genre' для каждого трека из этого датасета.
- **sample_submit.csv** - файл предсказаний в правильном формате.
  - instance_id - идентификатор трека в тестовом наборе.
  - music_genre - Целевой признак. Для каждого трека требуется категориальное значение соответствующее музыкальному жанру трека.
  
  
**Описание полей данных:** 
* **instance_id** -Уникальный идентификатор трека;
* **track_name** - Название трека;
* **acousticness** - Мера уверенности от 0,0 до 1,0 в том, что трек является акустическим. 1,0 означает высокую степень уверенности в том, что трек является акустическим;
* **danceability** - Танцевальность описывает, насколько трек подходит для танцев, основываясь на сочетании музыкальных элементов, включая темп, стабильность ритма, силу ударов и общую регулярность. Значение 0,0 означает наименьшую танцевальность, а 1,0 - наибольшую танцевальность;
* **duration_ms** - Продолжительность трека в миллисекундах;
* **energy** - Энергия это показатель от 0,0 до 1,0, представляющий собой меру интенсивности и активности. Как правило, энергичные композиции ощущаются как быстрые, громкие и шумные. Например, дэт-метал обладает высокой энергией, в то время как прелюдия Баха имеет низкую оценку этого параметра;
* **instrumentalness** - Определяет, содержит ли трек вокал. Звуки "Ooh" и "aah" в данном контексте рассматриваются как инструментальные. Рэп или разговорные треки явно являются "вокальными". Чем ближе значение инструментальности к 1,0, тем больше вероятность того, что трек не содержит вокала;
* **key** - базовый ключ (нота) произведения;
* **liveness** - Определяет присутствие аудитории в записи. Более высокие значения liveness означают увеличение вероятности того, что трек был исполнен вживую. Значение выше 0,8 обеспечивает высокую вероятность того, что трек исполняется вживую;
* **loudness** - Общая громкость трека в децибелах (дБ);
* **mode** - Указывает на модальность (мажорную или минорную) трека;
* **speechiness** - Речевой характер определяет наличие в треке разговорной речи. Чем более исключительно речевой характер носит запись (например, ток-шоу, аудиокнига, поэзия), тем ближе значение атрибута к 1,0. Значения выше 0,66 характеризуют треки, которые, вероятно, полностью состоят из разговорной речи. Значения от 0,33 до 0,66 характеризуют треки, которые могут содержать как музыку, так и речь, как в виде фрагментов, так и в виде слоев, включая такие случаи, как рэп-музыка. Значения ниже 0,33, скорее всего, представляют музыку и другие неречевые треки;
* **tempo** - Темп трека в ударах в минуту (BPM). В музыкальной терминологии темп представляет собой скорость или темп данного произведения и напрямую зависит от средней продолжительности тактов;
* **obtained_date** - дата загрузки в сервис;
* **valence** - Показатель от 0,0 до 1,0, характеризующий музыкальный позитив, передаваемый треком. Композиции с высокой валентностью звучат более позитивно (например, радостно, весело, эйфорично), а композиции с низкой валентностью - более негативно (например, грустно, депрессивно, сердито);
* **music_genre** - Музыкальный жанр трека.

```python
#напишем функцию для определения и замены выбросов
def outliers(row):
    q_1 = np.quantile(row, 0.25) 
    q_3 = np.quantile(row, 0.75) 
    IQR = q_3 - q_1
    return np.where((row < (q_1 - IQR * 1.5)) | (row > (q_3 + IQR * 1.5)), row.median(), row)
```

```python
data_train['duration_ms'] = outliers(data_train['duration_ms'])
```

```python
data_test['duration_ms'] = outliers(data_test['duration_ms'])
```

```python
sns.boxplot(x=data_train['duration_ms'])
plt.title('Длительность. Boxplot. Тренировочный датасет')
```

```python
sns.boxplot(x=data_test['duration_ms'])
plt.title('Длительность. Boxplot. Тестовый датасет')
```

### 2.3 Работа с пропусками

```python
data_train.isna().sum()
```

```python
data_test.isna().sum()
```

**Пропуски в двух датасетах присутствуют в столбцах *key*, *mode*, *tempo*.**

**Для начала заполним пропуски в тренировочном датасете:**

**Для каждого жанра найдем самые часто встречающиеся ноту, модальность и среднее значение темпа.**

**И заменим ими пропущенные значения.**

```python
for genre in data_train['music_genre'].unique():
    popular_key_rating = data_train[data_train['music_genre'] == genre]['key'].value_counts() 
    popular_key = popular_key_rating.index[0]
    data_train.loc[(data_train['music_genre'] == genre) & (data_train['key'].isna()), 'key'] = popular_key
    print(genre, ':', popular_key)
```

```python
for genre in data_train['music_genre'].unique():
    popular_mode_rating = data_train[data_train['music_genre'] == genre]['mode'].value_counts() 
    popular_mode = popular_mode_rating.index[0]
    data_train.loc[(data_train['music_genre'] == genre) & (data_train['mode'].isna()), 'mode'] = popular_mode
    print(genre, ':', popular_mode)
```

```python
for genre in data_train['music_genre'].unique():
    tempo_mean = data_train[data_train['music_genre'] == genre]['tempo'].mean()
    data_train.loc[(data_train['music_genre'] == genre) & (data_train['tempo'].isna()), 'tempo'] = round(tempo_mean, 3)
    print(genre, ':', round(tempo_mean, 3))
```

```python
data_train.isna().sum()
```

**C пропусками в тестовом датасете уже не все так просто. У нас нет колонки с музыкальным жанром, чтобы на него ориентироваться, поэтому заполним пропуски наиболее часто встречающимися значениями нот и модальности, а также средним значением темпа.**

```python
popular_key_rating = data_test['key'].value_counts() 
popular_key = popular_key_rating.index[0]
data_test.loc[(data_test['key'].isna()), 'key'] = popular_key
print('Самая распространенная нота:', popular_key)
```

```python
popular_mode_rating = data_test['mode'].value_counts() 
popular_mode = popular_mode_rating.index[0]
data_test.loc[(data_test['mode'].isna()), 'mode'] = popular_mode
print('Самая распространенная модальность:', popular_mode)
```

```python
tempo_mean = data_test['tempo'].mean()
data_test.loc[(data_test['tempo'].isna()), 'tempo'] = round(tempo_mean, 3)
print('Среднее значение темпа:', round(tempo_mean, 3))
```

```python
data_test.isna().sum()
```

**Все пропуски обработаны.**

**Можно заметить, что некоторых классов больше, чем других. Например, композиций Blues сильно больше чем композиций Hip-Hop.**

**А также применим метод upsampling для того, чтобы чуть больше выравнять баланс классов до анализа.**

```python
rat_hiphop = len(data_train_cleaned.loc[data_train_cleaned['music_genre']=="Blues"])//len(data_train_cleaned.loc[data_train_cleaned['music_genre']=="Hip-Hop"])
rat_jazz = len(data_train_cleaned.loc[data_train_cleaned['music_genre']=="Blues"])//len(data_train_cleaned.loc[data_train_cleaned['music_genre']=="Jazz"])
rat_classical = len(data_train_cleaned.loc[data_train_cleaned['music_genre']=="Blues"])//len(data_train_cleaned.loc[data_train_cleaned['music_genre']=="Classical"])
```

```python
df_hiphop = data_train_cleaned.loc[data_train_cleaned['music_genre']=="Hip-Hop"]
df_hiphop = df_hiphop.loc[df_hiphop.index.repeat(rat_hiphop)]

df_jazz = data_train_cleaned.loc[data_train_cleaned['music_genre']=="Jazz"]
df_jazz = df_jazz.loc[df_jazz.index.repeat(rat_jazz)]

df_classical = data_train_cleaned.loc[data_train_cleaned['music_genre']=="Classical"]
df_classical = df_classical.loc[df_classical.index.repeat(rat_classical)]

data_train_cleaned = pd.concat([data_train_cleaned.loc[data_train_cleaned['music_genre']!= "Hip-Hop"], df_hiphop]).sample(frac=1)
data_train_cleaned = pd.concat([data_train_cleaned.loc[data_train_cleaned['music_genre']!= "Jazz"], df_jazz]).sample(frac=1)
data_train_cleaned = pd.concat([data_train_cleaned.loc[data_train_cleaned['music_genre']!= "Classical"], df_classical]).sample(frac=1)
```

## 7. Выбор и обучение моделей

<a id='7.1'></a>

### 7.1 Формулировка ML-задачи 

**Главной задачей нашей модели является предсказание музыкальных жанров. Для оценки качества нам необходимо учесть и точность Precision, и полноту Recall, так как именно введение precision не позволяет нам записывать все объекты в один класс, так как в этом случае мы получаем рост уровня False Positive. Recall демонстрирует способность алгоритма обнаруживать данный класс вообще, а precision — способность отличать этот класс от других классов.**

**Часто в реальной практике стоит задача найти оптимальный (для заказчика) баланс между этими двумя метриками. F-мера — среднее гармоническое precision и recall.**

**Метрика F1-micro представляет собой обычную формулу для F1, но рассчитывается с использованием общего количества истинных положительных результатов (TP), ложных положительных результатов (FP) и ложных отрицательных результатов (FN), а не индивидуально для каждого класса.F1-micro придает одинаковую важность каждому наблюдению.**

**В соответствии с желанием заказчика, мы будем использовать метрику F1-micro для оценки качества моделей.**

<a id='7.2'></a>

1. Требуется введение дополнительных признаков для лучшего различия жанров **Alternative**, **Rap** и **Hip-Hop**.
2. В столбце **duration_ms** выявлены аномальные значения, равные **-1**. Стоит разобраться в их природе.
